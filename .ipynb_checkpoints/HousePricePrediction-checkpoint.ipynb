```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset: house sizes in 1000s sqft and prices in 1000s USD
X = np.array([[1], [2], [3], [4], [5]]) # Sizes
y = np.array([150, 200, 250, 300, 350]) # Prices

# Train the model
model = LinearRegression()
model.fit(X, y)

def predict_and_visualize():
    # Get user input
    try:
        house_size = float(input("Enter house size (in 1000s sqft, e.g., 6.5 for 6500 sqft): "))
    except ValueError:
        print("Please enter a valid number!")
        return
    
    # Make prediction
    predicted_price = model.predict([[house_size]])
    
    # Create visualization
    plt.figure(figsize=(12, 8))
    
    # Generate range for regression line
    X_range = np.linspace(0, max(10, house_size + 2), 100).reshape(-1, 1)
    y_pred_range = model.predict(X_range)
    
    # Plot training data
    plt.scatter(X, y, color='blue', s=80, label='Training Data', alpha=0.7, zorder=3)
    
    # Plot regression line
    plt.plot(X_range, y_pred_range, color='red', linewidth=3, 
             label=f'Regression Line: y = {model.coef_[0]:.1f}x + {model.intercept_:.1f}')
    
    # Plot the user's prediction point
    plt.scatter([house_size], predicted_price, color='green', s=200, 
                marker='*', label=f'Prediction: {house_size}k sqft = ${predicted_price[0]:.1f}k',
                edgecolors='black', zorder=5)
    
    # Add connecting lines to axes
    plt.axhline(y=predicted_price, color='green', linestyle='--', alpha=0.5)
    plt.axvline(x=house_size, color='green', linestyle='--', alpha=0.5)
    
    # Customize the plot
    plt.xlabel('House Size (1000s sqft)', fontsize=12)
    plt.ylabel('Price (1000s USD)', fontsize=12)
    plt.title(f'House Price Prediction: {house_size}k sqft = ${predicted_price[0]:.1f}k', 
              fontsize=14, fontweight='bold')
    
    # Add grid and legend
    plt.grid(True, alpha=0.3)
    plt.legend()
    
    # Add information box
    info_text = f'''Model Information:
Slope: ${model.coef_[0]:.1f}k per 1000 sqft
Intercept: ${model.intercept_:.1f}k
R¬≤ Score: {model.score(X, y):.3f}
Prediction: {house_size}k sqft = ${predicted_price[0]:.1f}k'''
    
    plt.text(0.02, 0.98, info_text, transform=plt.gca().transAxes, 
             fontsize=10, verticalalignment='top',
             bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    # Print results
    print(f"\n{'='*50}")
    print(f"PREDICTION RESULT:")
    print(f"{'='*50}")
    print(f"Input house size: {house_size:,} sqft")
    print(f"Predicted price: ${predicted_price[0]:,.2f}")
    print(f"Model confidence (R¬≤): {model.score(X, y):.3f}")
    print(f"{'='*50}")

# Interactive loop
def main():
    print(" HOUSE PRICE PREDICTION SYSTEM ")
    print("=" * 40)
    
    while True:
        print("\nOptions:")
        print("1. Make a prediction")
        print("2. Exit")
        
        choice = input("\nEnter your choice (1 or 2): ").strip()
        
        if choice == '1':
            predict_and_visualize()
        elif choice == '2':
            print("Thank you for using the prediction system!")
            break
        else:
            print("Invalid choice! Please enter 1 or 2.")

# Run the program
if __name__ == "__main__":
    main()
```
